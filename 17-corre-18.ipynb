# Correlation
- relation of the variables in the dataset
  
**TYPES**
- pearson`s
- sprearman
- kendall`s
  
 **Example**
- one variable depend on the other variable
- One variable could slightly depend on other variable
- Two varibles depend on the third variable

1-Positive correlation
2-Negative correlation
3-No relation

```python
import seaborn as sns 
import pandas  as pd
import numpy as np
```
